### Gera arquivo no formato banco de dados (cada linha é uma data)

```python
# Importando os pacotes 
import pandas as pd
import numpy as np
import csv
import os
from os import listdir
from os.path import isfile, join
```

```python
# Diretorio
diretorio="/home/gma/Dropbox/Public/AgriBIsiness/DadosMF"

with open('20190718 MFBGNN.csv', encoding='latin-1') as f:
    reader = csv.reader(f, delimiter='\t')
    for row in reader:
        print(row)
```

```python
# Diretorio
diretorio="/home/gma/Dropbox/Public/AgriBIsiness/DadosMF"

# Recuperando lista de ficheiros CSV do diretorio
ficheiros = [f for f in listdir(diretorio) if (isfile(join(diretorio, f)) and f.endswith('.csv')) ]

datas = []
Pracas = []
df_valores = pd.DataFrame()

# Recuperando as datas e valores para cada ficheiro
for item in ficheiros:
    vard = item[0:8]
    data = int(vard)
    datas.append(data)
    preco = []
    with open(item, encoding='latin-1') as f:
        reader = csv.reader(f, delimiter=';')
        for row in reader:
            valor = row[1]
            praca = row[0]
            preco.append(valor)
            Pracas.append(praca)
            Pracas = Pracas[0:32]                            
    df = pd.DataFrame(preco)
    dft = df.T
    df_valores = pd.concat([df_valores, dft], ignore_index=True)
df_valores.columns = Pracas
df_datas = pd.concat([pd.DataFrame([i], columns=['Data']) for i in datas], ignore_index=True)
```

```python
# Unindo datas e valores num unico dataframe
df_fim = df_datas.join(df_valores)
df_fim
```

```python
# Salvando dataframe como arquivo '.csv'
df_fim.to_csv("teste1.csv")
```

```python
# Movendo banco de dados gerado para outro diretório
path = "/home/gma/Dropbox/Public/AgriBIsiness/DadosMF/teste1.csv"
dst = "/home/gma/Dropbox/Public/AgriBIsiness/teste1.csv"
os.link(path,dst)
os.remove("/home/gma/Dropbox/Public/AgriBIsiness/DadosMF/teste1.csv")
```
